## **Data Visualization**

1.What is the typical session duration for users, and are there any significant spikes in very short or long sessions?

```python
plt.hist(df['Session Duration'], bins=15, color='orange')
plt.title('Session Duration')
plt.show()
```

2.Which traffic sources lead to the highest or most consistent conversion rates, and are there any sources with a wide range of outcomes?

```python
df.boxplot(column='Conversion Rate', by='Traffic Source',color='orange')
plt.title('Conversion Rate by Source')
plt.suptitle('')
plt.xticks(rotation=45)
plt.show()
```

3.Is there a relationship between how long a user spends on a page and their likelihood of converting?

```python
plt.scatter(df['Time on Page'], df['Conversion Rate'], color='orange',alpha=0.7)
plt.title('Time on Page vs Conversion Rate')
plt.show()
```

4.Which traffic source leads to the highest average conversion rate, and how do the others compare in performance?

```python
avg_conversion = df.groupby('Traffic Source')['Conversion Rate'].mean()
avg_conversion.plot(kind='line', marker='o', color='orange', title='Avg Conversion Rate by Source')
plt.xlabel('Traffic Source')
plt.ylabel('Average Conversion Rate')
plt.grid(True)
plt.show()
```

What proportion of website traffic is coming from each source, and are certain channels significantly more dominant than others

```python
df['Traffic Source'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Website Visitors by Traffic Source')
plt.ylabel('')  # Hides the y-axis label for a cleaner look
plt.show()
```
